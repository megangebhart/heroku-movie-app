```python
# Import Dependencies

import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import confusion_matrix

import numpy as np
import itertools
import matplotlib.pyplot as plt

import pandas as pd
from glob import glob
import os 
```

```python
# TRAINING DATASET
# Defining path for negative reviews
path = os.path.join("Dataset","train","neg")


# Getting file names 
all_files = glob(path + "\*.txt")

# List for saving the files text
li = []
labels = []

# Looping for getting the reviews 
for filename in all_files:
    try:
        read_f = open(filename,'r')
        for line in read_f:
            li.append(line)
            labels.append(1)
        read_f.close()
    except UnicodeDecodeError:
        print("There was an error here! :(")

# Reading positive reviews files 
path = os.path.join("Dataset","train","pos")
# Getting file names 
pos_files = glob(path + "\*.txt")     
        
# Looping for getting the reviews 
for filename in pos_files:
    try:
        read_f = open(filename,'r')
        for line in read_f:
            li.append(line)
            labels.append(0)
        read_f.close()
    except UnicodeDecodeError:
        print("There was an error here! :(")
        
        
review_train_df = pd.DataFrame()
review_train_df["reviews"]=li
# Negative label = 1 - Positive label = 0
review_train_df["label"] = labels
review_train_df
```

```python
# TESTING DATASET
# Defining path for negative reviews
path = os.path.join("Dataset","test","neg")

# Getting file names 
all_files = glob(path + "\*.txt")

# List for saving the files text
li = []
labels = []

# Looping for getting the reviews 
for filename in all_files:
    try:
        read_f = open(filename,'r')
        for line in read_f:
            li.append(line)
            labels.append(1)
        read_f.close()
    except UnicodeDecodeError:
        print("There was an error here! :(")

# Reading positive reviews files 
path = os.path.join("Dataset","test","pos")
# Getting file names 
pos_files = glob(path + "\*.txt")     
        
# Looping for getting the reviews 
for filename in pos_files:
    try:
        read_f = open(filename,'r')
        for line in read_f:
            li.append(line)
            labels.append(0)
        read_f.close()
    except UnicodeDecodeError:
        print("There was an error here! :(")
        
        
review_test_df = pd.DataFrame()
review_test_df["reviews"]=li
# Negative label = 1 - Positive label = 0
review_test_df["label"] = labels
review_test_df
```
